**chart7:line plot for
checking the effect of avg dailly rate on cancelation**
#the plt for daily avg rate for cancelled and non cancelled is prepared

```python
canceled_adr2=canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_adr2.reset_index(inplace=True)
canceled_adr2.sort_values('reservation_status_date',inplace=True)

non_caceled_adr=data[data['is_canceled']==0]
non_caceled_adr=non_caceled_adr.groupby('reservation_status_date')[['adr']].mean()
non_caceled_adr.reset_index(inplace=True)
non_caceled_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(16,6))
plt.title('average daily rate')
plt.plot(canceled_adr2['reservation_status_date'],canceled_adr2['adr'],label='cancelled',color='red')

plt.plot(non_caceled_adr['reservation_status_date'],non_caceled_adr['adr'],label='non cancelled',color='blue')
plt.title('cancelled adr vs non cancelled adr' )
plt.show()
```
